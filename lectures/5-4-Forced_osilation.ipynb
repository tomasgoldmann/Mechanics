# Forced oscilation

**Forced oscillation** is a type of mechanical oscillation that occurs when an external force acts on an oscillating system. Unlike free oscillation, where the system oscillates only under the influence of initial conditions, in forced oscillation there is an external excitation that maintains the motion.

## Harmonic excitation

The most commonly considered type of excitation is the **harmonic force**, which has the form of a sine or cosine function of time:

$$
F(t) = F_0 \cos(\omega t)
$$

where:
- $F_0$ is the amplitude of the excitation force,
- $\omega$ is the angular frequency of the excitation,
- $t$ is the time.

This excitation occurs, for example, in rotating machines, where the force is applied periodically.

## Equation of motion for undamped harmonic oscillation

Consider a mechanical oscillator of mass \(m\) that is connected to a spring of stiffness \(k\) and subjected to an external harmonic force \(F(t)\). The equation of motion is based on Newton's second law:

$$
m \ddot{x} = -k x + F_0 \cos(\omega t)
$$

After modification, we obtain the differential equation of forced oscillation:

$$
m \ddot{x} + kx = F_0 \cos(\omega t)
$$

The natural angular frequency of the oscillator is often introduced:

$$
\omega_0 = \sqrt{\frac{k}{m}}
$$

and the equation is rewritten in the form:

$$
\ddot{x} + \omega_0^2 x = \frac{F_0}{m} \cos(\omega t)
$$

This equation describes forced undamped oscillation under the influence of harmonic excitation.

### Solution of the equation of motion for undamped harmonic oscillation

The general solution of the equation of motion consists of two parts:
1. **General solution of the homogeneous equation** (corresponds to free oscillation):

$$
x_h(t) = C_1 \cos(\omega_0 t) + C_2 \sin(\omega_0 t)
$$

where $C_1$ and $C_2$ are the integration constants determined from the initial conditions.

2. **Particular solution** (corresponds to forced oscillation caused by an external force):

We are looking for a particular solution in the form:

$$
x_p(t) = A \cos(\omega t) + B \sin(\omega t)
$$

Substituting into the differential equation:

$$
-A \omega^2 \cos(\omega t) - B \omega^2 \sin(\omega t) + \omega_0^2 (A \cos(\omega t) + B \sin(\omega t)) = \frac{F_0}{m} \cos(\omega t)
$$

Comparing the coefficients of \(\cos(\omega t)\) and \(\sin(\omega t)\) we get the system:

$$
(\omega_0^2 - \omega^2) A = \frac{F_0}{m}, \quad (\omega_0^2 - \omega^2) B = 0
$$

From the second the equation follows \( B = 0 \), and therefore:

$$
A = \frac{F_0 / m}{\omega_0^2 - \omega^2}
$$

assuming that $$(\omega \neq \omega_0)$$ (i.e. we are not in resonance).

### General solution

The general solution of the equation of motion is given by the sum of the homogeneous and particular solutions:

$$
x(t) = C_1 \cos(\omega_0 t) + C_2 \sin(\omega_0 t) + \frac{F_0 / m}{\omega_0^2 - \omega^2} \cos(\omega t).
$$

The first two terms correspond to a transient state that disappears after a certain time, while the last term describes a steady forced oscillation.

## Resonance

> **Resonance** is a phenomenon that occurs when the excitation frequency $\omega$ of the external force acting on the oscillator is equal to the natural frequency of the system $\omega_0$. In this case, the amplitude of the oscillations increases sharply.

If $\omega \to \omega_0$, the denominator in the expression for the amplitude of the particular solution $\frac{F_0 / m}{\omega_0^2 - \omega^2}$ approaches zero, which means that the amplitude of the oscillations theoretically approaches infinity. This phenomenon is called **resonance** and in real systems is limited by damping.

### Simple derivation of resonance

Assume that both the force and the oscillation are in phase and the oscillation of the oscillator can be described by the equation

$$x(t)= A \cos \omega t$$

substituting into the equation of motion we obtain

$$
\ddot{x} + \omega_0^2 x = \frac{F_0}{m} \cos(\omega_0 t)
$$

$$
- A \omega^2 \cos \omega t + A \omega_0^2 \cos \omega t = \frac{F_0}{m} \cos(\omega_0 t)
$$

for the amplitude of the oscillations in resonance we have

$$
A = \frac{F_0}{m} \frac{1}{\omega^2 - \omega_0^2}
$$

## Damped Forced Oscillations

During damped forced oscillations, three types of forces act on the oscillator:

1. **Restoring force** of the spring according to Hooke's law: $F_k = - k x$.
2. **Damping force** proportional to the speed: $F_t = - b \dot{x}$.
3. **Exciting force** of harmonic nature: $F = F_0 \cos(\omega t)$.

Newton's second law gives the equation of motion:

$$
m \ddot{x} + b \dot{x} + k x = F_0 \cos(\omega t).
$$

We introduce:

- **Eigencircular frequency** $\omega_0 = \sqrt{\frac{k}{m}}$,
- **Damping coefficient** $\zeta = \frac{b}{2m\omega_0}$,
- **Exciting force per unit mass** $f = \frac{F_0}{m}$.

The equation of motion is then rewritten in the form:

$$
\ddot{x} + 2\zeta\omega_0 \dot{x} + \omega_0^2 x = f \cos(\omega t).
$$

### Solution of the equation

The total solution $x(t)$ consists of two components:

1. **Homogeneous solution** $x_h(t)$ corresponding to free damped oscillation.
2. **Particular solution** $x_p(t)$, which corresponds to steady forced motion.

### Homogeneous solution

We solve the equation:

$$
\ddot{x} + 2\zeta\omega_0 \dot{x} + \omega_0^2 x = 0.
$$

Characteristic equation:

$$
\lambda^2 + 2\zeta\omega_0 \lambda + \omega_0^2 = 0.
$$

The solution depends on the size of $\zeta$:

- **Subcritical damping** $\zeta < 1$: oscillations with exponential damping,

$$
x_h(t) = e^{-\zeta \omega_0 t} (C_1 \cos(\omega_d t) + C_2 \sin(\omega_d t)),
$$

or
$$
x_h(t) = e^{-\zeta \omega_0 t} C \cos(\omega_d t + \psi),
$$

where $$\omega_d = \omega_0 \sqrt{1 - \zeta^2}$$ is the damped frequency.

- **Critical and supercritical damping**: non-oscillating motion.

### Particular solution

We are looking for in the form:

$$
x_p(t) = A \cos(\omega t) + B \sin(\omega t).
$$

First derivative:

$$
\dot{x}_p(t) = -A \omega \sin(\omega t) + B \omega \cos(\omega t)
$$

Second derivative:

$$
\ddot{x}_p(t) = -A \omega^2 \cos(\omega t) - B \omega^2 \sin(\omega t).
$$

Now we substitute $x_p(t)$, $\dot{x}_p(t)$ and $\ddot{x}_p(t)$ into the equation:

$$
(-A \omega^2 \cos(\omega t) - B \omega^2 \sin(\omega t))
+ 2\zeta\omega_0 (-A \omega \sin(\omega t) + B \omega \cos(\omega t))
+ \omega_0^2 (A \cos(\omega t) + B \sin(\omega t))
= f \cos(\omega t).
$$

We will list the members:

$$
(-A \omega^2 + A \omega_0^2) \cos(\omega t) + (-B \omega^2 + B \omega_0^2) \sin(\omega t)
+ (-2\zeta\omega_0 A \omega) \sin(\omega t) + (2\zeta\omega_0 B \omega) \cos(\omega t)
= f \cos(\omega t).
$$

We group terms according to the functions $\cos(\omega t)$ and $\sin(\omega t)$:

$$
(A (\omega_0^2 - \omega^2) + 2\zeta\omega_0 B \omega) \cos(\omega t)
+ (B (\omega_0^2 - \omega^2) - 2\zeta\omega_0 A \omega) \sin(\omega t)
= f \cos(\omega t).
$$

Since the equation must hold for all \( t \), we compare the coefficients of \(\cos(\omega t)\) and \(\sin(\omega t)\):

1. For $\cos(\omega t)$:

$$
A (\omega_0^2 - \omega^2) + 2\zeta\omega_0 B \omega = f
$$

3. For $\sin(\omega t)$:

$$
B (\omega_0^2 - \omega^2) - 2\zeta\omega_0 A \omega = 0
$$

From the second equation, we express $B$:

$$
B = \frac{2\zeta\omega_0 A \omega}{\omega_0^2 - \omega^2}
$$

Substitute into the first equation:

$$
A (\omega_0^2 - \omega^2) + 2\zeta\omega_0 \cdot \frac{2\zeta\omega_0 A \omega}{\omega_0^2 - \omega^2} \omega = f
$$

We will subtract $A$:

$$
A \left( (\omega_0^2 - \omega^2) + \frac{4\zeta^2\omega_0^2\omega^2}{\omega_0^2 - \omega^2} \right) = f
$$

We put the common denominator:

$$
A \frac{(\omega_0^2 - \omega^2)^2 + (2\zeta\omega_0\omega)^2}{\omega_0^2 - \omega^2} = f
$$

From here we express $A$:

$$
A = \frac{f (\omega_0^2 - \omega^2)}{D},
$$

where

$$
D = (\omega_0^2 - \omega^2)^2 + (2\zeta\omega_0\omega)^2.
$$

Similarly for $B$:

$$
B = \frac{f (2\zeta\omega_0\omega)}{D}.
$$

Total amplitude:

$$
X = \frac{f}{\sqrt{(\omega_0^2 - \omega^2)^2 + (2\zeta\omega_0\omega)^2}}.
$$

Phase shift:

$$
\tan \phi = \frac{2\zeta\omega_0\omega}{\omega_0^2 - \omega^2}.
$$

### Overall solution

The overall solution is:

$$
x(t) = e^{-\zeta \omega_0 t} (C_1 \cos(\omega_d t) + C_2 \sin(\omega_d t)) + X \cos(\omega t - \phi).
$$

In the steady state, the forced component $X \cos(\omega t - \phi)$ dominates. At resonance $(\omega \approx \omega_0)$ the amplitude increases to a maximum, but is not infinite due to damping.
