## Strain energy and its density under spatial stress

## Definition of strain energy

> **Strain energy** is the energy that accumulates in a body as a result of its elastic deformation under the action of external forces.

This energy is **reversible** in the elastic behavior of a body – it is released when the load is removed and the body returns to its original state.

## Strain energy density

### Strain energy density for uniaxial stress

Assume that we have a force $F$ that axially loads a rod and causes a deformation of $\Delta l$. The work done is the elastic deformation

$$W = \frac{1}{2} F \Delta l $$

which by dividing by the volume we can get

$$w = \frac{W}{V} = \frac{1}{2} \frac{F}{A} \frac{\Delta l}{l} = \frac{1}{2} \sigma \varepsilon$$

> **Strain energy density** ($w$) is the strain energy per unit volume of the material.

### Strain energy density for shear loading

Suppose we have a force $T$ that axially loads the material in shear and causes a strain of $\Delta x$. The work done is the work of elastic deformation

$$W = \frac{1}{2} T \Delta x $$

which by dividing by the volume we can get

$$w = \frac{W}{V} = \frac{1}{2} \frac{F}{A} \frac{\Delta x}{y} = \frac{1}{2} \tau \gamma$$

### Strain energy density for general stress

For the case of a **linearly elastic material** in **3D stress** it can be expressed as:

$$
w = \frac{1}{2} (\sigma_x \varepsilon_x + \sigma_y \varepsilon_y + \sigma_z \varepsilon_z + \tau_{xy} \gamma_{xy} + \tau_{yz} \gamma_{yz} + \tau_{zx} \gamma_{zx})
$$

where:
- $\sigma_x, \sigma_y, \sigma_z$ are the principal stress components,
- $\varepsilon_x, \varepsilon_y, \varepsilon_z$ are the principal strain components,
- $\tau_{xy}, \tau_{yz}, \tau_{zx}$ are the shear stresses,
- $\gamma_{xy}, \gamma_{yz}, \gamma_{zx}$ are the corresponding shear strains.

### Derivation from generalized Hooke's law

In general tensor notation (stress and strain as second-order tensors) we have:

$$
w = \frac{1}{2} \, \boldsymbol{\sigma} : \boldsymbol{\varepsilon}
$$

where the operation $:$ denotes **double scalar product of tensors**, i.e.:

$$
\boldsymbol{\sigma} : \boldsymbol{\varepsilon} = \sum_{i=1}^{3} \sum_{j=1}^{3} \sigma_{ij} \varepsilon_{ij}
$$

### Alternative expression for isotropic material

Using the stress-strain relations from linear elasticity (Hooke's law for isotropic materials), the strain energy density can also be expressed in terms of stress only:

$$
w = \frac{1 + \nu}{2E} (\sigma_x^2 + \sigma_y^2 + \sigma_z^2) - \frac{\nu}{E} (\sigma_x \sigma_y + \sigma_y \sigma_z + \sigma_z \sigma_x) + \frac{1}{2G} (\tau_{xy}^2 + \tau_{yz}^2 + \tau_{zx}^2)
$$

where:
- $E$ is Young's modulus of elasticity,
- $\nu$ is Poisson's ratio,
- $G$ is the shear modulus of elasticity.

### Meaning of strain energy

The strain energy density is used:
- in the calculation of internal forces in the finite element method (FEM),
- in the calculation of the failure criterion (e.g. **failure energy criterion - von Mises**),
- in determining the area of ​​plastic deformation,
- in the design of materials and structures that are to absorb energy (e.g. protective elements, bumpers).
